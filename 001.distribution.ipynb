```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

n, p = 10, 0.5
x = np.arange(binom.ppf(0.01, n, p),binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

ax.legend(loc='best', frameon=False)
ax.set_xlabel('表の枚数')
ax.set_ylabel('確率')
ax.set_title('二項分布(10回コインを投げる)')
plt.show()
```

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

n, p = 100, 0.5
x = np.arange(binom.ppf(0.01, n, p),binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='普通のコイン')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)

n, p = 100, 0.3
x = np.arange(binom.ppf(0.01, n, p),binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'go', ms=8, label='イカサマコイン')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='g', lw=5, alpha=0.5)

ax.legend(loc='best', frameon=False)
ax.set_xlabel('表の枚数')
ax.set_ylabel('確率')
ax.set_title('二項分布(100回コインを投げる)')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),'r-', lw=2, alpha=0.5, label='norm pdf')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
df = 10
x = np.linspace(t.ppf(0.01, df),t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df),'g-', lw=2, alpha=0.5, label='t pdf (df=10)')

df2 = 5
x2 = np.linspace(t.ppf(0.01, df2),t.ppf(0.99, df2), 100)
ax.plot(x2, t.pdf(x2, 1),'b-', lw=2, alpha=0.5, label='t pdf (df=5)')

x3 = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
ax.plot(x3, norm.pdf(x3),'r-', lw=2, alpha=0.5, label='norm pdf')

ax.legend(loc='best', frameon=False)
plt.show()
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
    
fig, ax = plt.subplots(1, 1)
df = 1
x = np.linspace(chi2.ppf(0.01, df),chi2.ppf(0.99, df), 100)
ax.plot(x, chi2.pdf(x, df),'r-', lw=2, alpha=0.5, label='chi2 pdf (df=1)')

df2 = 2
x = np.linspace(chi2.ppf(0.01, df2),chi2.ppf(0.99, df2), 100)
ax.plot(x, chi2.pdf(x, df2),'b-', lw=2, alpha=0.5, label='chi2 pdf (df=2)')

ax.legend(loc='best', frameon=False)
plt.show()
```
